### Brief description of the objects:

Our observation targets consist of two main classes of objects:
1. One **supernova remnant (SNR)**: the first target on the list is a remnant of a supernova explosion within the Milky Way. Despite being within our own galaxy, the visual magnitude is quite high, indicating a faint source, which might be a challenge to observe.
2. **Type-1 active galactic nuclei (AGNs)**: the rest of our targets are type 1 AGNs, most of which are also Seyfert 1 galaxies. These are galaxies with a luminous cores powered by accretion of matter onto a central supermassive black hole, specifically with both broad and narrow emission lines in their spectra. Such objects are extragalactic targets with a variety of visual magnitudes, spanning from relatively bright (e.g. NGC3227) to faint (e.g. Mrk110).

### Calculating the Field-of-View (FOV)

First, we calculate the size of the FOV of the Milankovic telescope, using the following CCD camera in port 1 which is designed for scientific imaging.

This port is equipped with focal reducer 0.65, so focal length is 7132 mm (instead of 11200 mm). There is also field flattener f/7.6 which gives F=10.64 m.

**Andor iKon-L 936 CCD**<br>
Andors iKon-L 936 is designed with scientific imaging in mind. The 2048 x 2048 array and 13.5 m pixel pitch combine to deliver a 27.6 x 27.6 mm active image area, TE cooled down to -100C. The iKon-L offers outstanding resolution, field of view, sensitivity and dynamic range performance. Ultimate sensitivity performance is achieved through combination of > 90% QE (back-illuminated sensor), low noise readout electronics and exceptionally deep TE cooling.

**Active pixels:** 					2048 x 2048<br>
**Sensor size:**					27.6 x 27.6 mm<br>
**Pixel size (W x H):**				13.5 m x 13.5 m<br>
**Active area pixel well depth:** 	100,000 e- (150,000 e- for BEX2-DD and BR-DD models)<br>
**Maximum readout rate:**			5 MHz<br>
**Read noise:**						2.9 e-<br>
**Maximum cooling:**				-100C<br>

This camera is equiped with following filters: Johnson B, V, R, I, L, filters and H , H continuum and SII, which makes it suitable to observe our selected targets.

We use the following formula [9] to calculate the FOV:

$$P=\dfrac{206265k}{1000f}$$

in which we can input the following values:
- $F = 10.64$ m is the focal length in mm of the primary mirror with the focal reducer and field flattener.
- $k = 13.5$ m is the pixel size

to obtain the plate scale:
$$P[arcsec/pixel]=206265\times13.5/1000*10.64*1000=0.2617084117$$

Now, we compute the total size of the field of view by multiplying by the number of active pixels (2048):

$$\mathrm{FOV\:dimension}=P\times2048 =0.2617084117\times2048=535.9788272\:\mathrm{arcsec}=8.932980453\:\mathrm{arcmin}$$

So, our instrument has a 9 x 9 arcmin FOV.
