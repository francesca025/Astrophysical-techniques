The emission coming from the astronomical sources is affected by the matter present between the source and observer, mainly by the interstellar matter (dust and gas) and by the Earth's atmosphere. To minimize the effects of the atmosphere, professional observatories are built on high and dry places with lower precipitation and less cloudy nights. Another, more challenging and expensive, solution to this problem is to launch telescopes in space, outside the influence of the Earth's atmosphere.

The radiation of an astronomical object (e.g., a star) passing through the atmosphere is affected by refraction (which causes changes in the apparent position of a star) and atmospheric extinction, which attenuate the flux reaching the Earth's surface (denoted as e.g., $f^{A}_{\lambda}$, while the 
flux at the top of the atmosphere is denoted with $f_{\lambda}$); here we neglect the absorption by photometric filters). The atmosphere has less impact on objects closer to the zenith than on those near the horizon because the light passes through a shorter geometric path, encountering fewer atmospheric layers, and thus suffers less extinction. The parameter that describes the thickness of the atmosphere is defined as airmass X. A higher airmass indicates that the object is closer to the horizon. The airmass is normalized to unity at the zenith X = 1. This is an imaginary boundary above which we consider having a vacuum, with no atmospheric effect.

As the zenith distance increases, the airmass grows. Several models (for both plane-parallel and spherical atmospheres, see Mathar 2015) describe how airmass changes with the zenith distance of the object. In the first-order approximation for a spherical atmosphere, the expression is given by:

$$X=\dfrac{1}{\cos(Z)}\left(1-0.0012\left(\dfrac{1}{\cos^2(Z)}-1\right)\right)$$

where $X$ is the zenith distance of the object. This expression is better than the plane-parallel case because it can be used for larger zenith distances (the plane-parallel case is limited to $Z=70\degree$ or $X=3$).

## Additional task
Independently derive the expression for airmass in the case of the plane-parallel atmosphere using the extinction law (Beer-Lambert law).

Beer-Lambert law states that $I_z=I^0\exp^{-k F(z)}$, where:
- $I_z$ is the intensity of light observed at the Earth's surface (after passing through the atmosphere)
- $I^0$ is the intensity of light outside the Earth's atmosphere (before atmospheric extinction)
- $k$ is the extinction coefficient/optical depth per unit air mass
- and $F(z) = \int^d_0 \rho ds$ is the airmass function, so that we have:
$$I_{z=0}=I^0\exp^{-k F(0)}$$
$$\implies I_z=I_{z=0}\exp^{-k(F(z)-F(0))} $$
$$\implies I_z=I_{z=0}\exp^{-k'(X-1)}$$
such that $X=\dfrac{F(z)}{F(0)}$ gives us the airmass.

$$X=\dfrac{F(z)}{F(0)}F(z) = \dfrac{\int^d_0 \rho ds}{\int^0_0 \rho ds}$$

**Applying the plane-parallel approximation**, we assume the density of air to be uniform above the observer i.e. that $\rho$ is independent of $s$:

$$X=\dfrac{\rho\int^d_0 ds}{\rho\int^0_0 ds}=\dfrac{L_z}{L_0}$$

where $L_z$ is the distance to the atmosphere along some nonzero zenith angle $z$ (hypotenuse in the below geometry), and $L_0$ is the zenith height to the atmosphere (vertical distance in the above geometry).

![Figure_2_techniques_HW1.png](Figure_2_techniques_HW1.png)

From the geometry in the above figure:

$$L_z=\dfrac{L_0}{\cos(z)}$$

such that: $$X=\dfrac{L_0/\cos(z)}{L_0}=\dfrac{1}{\cos(z)}$$

The extinction in the atmosphere is cause by several mechanisms:
- Rayleigh scattering on molecules: this scattering is proportional to $\lambda^{-4}$, and is responsible for the blue color of the sky. It changes minimally over time.
- Ozone absorpotion: does not change drastically over short time intervals; defines the short-wavelength limit of atmospheric transparency.
- Aerosol scattering (Mie scattering): proportional to $\lambda^{-1}$, and is responsible for the white-gray color of clouds. It varies signidficantly over time and has the greatest influence up to altitudes of about 4.5 km.
- Molecular absorption: the main absorbers are water vapor, carbon dioxide, and oxygen. These vary considerably with time, location, and temperature.

Each of these processes depends on the wavelength of the radiation. Figure 1 illustrates the contribution of each mechanism to the total extinction curve for different wavelengths.

![Figure_1_techniques_HW1.png](Figure_1_techniques_HW1.png)

The magnitude observed at the Earth's surface can be described using the extinction (Beer-Lambert) law:

$$m^A_\lambda=m_\lambda+k_\lambda X$$

where $k_\lambda$ is the extinction coefficient for a given wavelength (or filter) and is assumed to remain constant over time and isotropic. Under these assumptions, we consider the atmosphere to be **photometric**. Based on this, professional observatories characterize if an observing night had photometric conditions, indicating most optimal observing conditions in terms of the atmosphere properties.

By observing a standard (or non-variable) star (or multiple stars) over the course of a single night at different airmasses, we can estimate the extinction coefficient and the magnitude outside the atmosphere using the above equation. In addition to the extinction coefficient that we can use to correct the magnitude, we can also obtain a correction coefficient for a given color index CI. However, this coefficient changes from star to star due to different pectrum 
slopes, requiring the introduction of a second-order extinction coefficient. The extinction coefficient for the color index is given by:

$$k_{\lambda_1-\lambda_2}=k'+k''\mathrm{CI}$$

where CI is the color index, and $k'$ and $k''$ are the first- and second-order extinction coefficients, respectively.

To find $Z$, we follow the computations in [2]. First, we need to find the coordinates of the stars by computing the Hour Angle:

$$H = \mathrm{LST} - \alpha$$

Note that currently, both the LST and $\alpha$ are in minutes.

```python
df['HA'] = df['LST'] - df['ra'] #find hour angle in minutes
df
```

Note that we must account for the times past midnight since the hour angle $H$ must fall in $[0, 24)$ hrs. We can do this by adding 24 hrs ($24\times 60$ minutes) to the computed $H$.

```python
for i in [3, 7, 11, 15, 19, 23]: #negative HA
    if df.loc[i, 'HA']<0:
        df.loc[i, 'HA']=df.loc[i, 'HA']+(24*60)
    else:
        print('Negative value at index ', i, ' has already been accounted for.')
```

```python
df['HA']
```

Now, we need to convert the hour angle into degrees. Following [7]:

1 Hour Angle [hours] = 60 Hour Angle [minutes] = $15\degree$

such that 1 Hour Angle [minutes] = $15/60\degree$.

```python
df['HA_deg'] = df['HA']*15/60
#df
```

Next, to calculate the zenith distance $Z$, we need to use the following formula from [2]:

$$\cos(Z)=\sin\phi\sin\delta+\cos\phi\cos\delta\cos H$$

where:
- $Z$ is the zenith distance
- $\phi$ is the latitude of the observer (Toronto observatory)
- $\delta$ is the star's declination
- $H$ is the hour angle in degrees

such that $Z$ can be found from: $Z=\arccos(\sin\phi\sin\delta+\cos\phi\cos\delta\cos H)$


```python
#to account for np trigonometric function in radians
rad_factor = np.pi/180
#print(1*rad_factor)
```

```python
#define a function to carry out the above computation
def find_Z(phi, delta, H):
    
    #account for functions in radians
    arg = np.sin(phi*rad_factor)*np.sin(delta*rad_factor) + np.cos(phi*rad_factor)*np.cos(delta*rad_factor)*np.cos(H*rad_factor)
    #print('Cos Z = ', arg)
    #print('Z = ', np.arccos(arg))
    return np.arccos(arg) #computation in radians
```

```python
#data for Toronto observatory
obs_phi = (43*60)+52 # latitude of observer (in min)

#convert this from minutes to degrees
obs_phi = obs_phi*15/60
```

```python
df['Z']=find_Z(obs_phi, df['dec'], df['HA_deg'])
df
```

Next, we calculate the air mass $X$ from the given relation:

$$X=\dfrac{1}{\cos(Z)}\left(1-0.0012\left(\dfrac{1}{\cos^2(Z)}-1\right)\right)$$

```python
#define a function to compute the airmass

def find_X(inp_z):
    # INPUTS: inp_z = zenith distance of object
    # OUTPUTS: X = air mass
    
    part_1=1/np.cos(inp_z*rad_factor)
    #print('1/cos(z)=', part_1)
    part_2 = 1 - 0.0012*(1/(np.cos(inp_z*rad_factor))**2-1)
    #print(part_2)
    return part_1 * part_2
```

```python
df['X'] = find_X(df['Z'])
df
```

Now, we fit the extinction (Beer-Lambert) law:

$$m^A_\lambda=m_\lambda+k_\lambda X$$

using the airmass $X$ and the relationship between the instrumental magnitudes in the V band $m^A_V$ and the magnitude of the star outside the atmosphere $m_V$ to find the respective extinction coefficient $k_V$.

```python
df['m_instr_V-m_V']=df['m_instr_V']-df['m_V'] #define Y values of line
# X values of line are X
```

```python
#define their relation (a straight line with 1 fit parameter)
def y(x, k):
    return k*x
```

```python
popt, pcov = curve_fit(y, df['X'], df['m_instr_V-m_V'])
```

```python
k_V = popt[0]
print('Slope: k=', k_V)
```

```python
x_fit = np.linspace(min(df['X']), max(df['X']), 100)  # 100 points from min(x) to max(x) for a smooth line
y_fit = y(x_fit, k_V)

plt.scatter(df['X'], df['m_instr_V-m_V'], label = 'Data', color='black', marker='+')
plt.plot(x_fit, y_fit, label = 'Fitted line', color='red', ls='--')
#plt.xlim(0.999, 1.001)
plt.title(r'Fitting to find $k_V$')
plt.xlabel(r'Airmass $X$')
plt.ylabel(r'$m^A_V-m_V$')
plt.legend()
```

**Result 1:** we have found the extinction coefficient in the V band to be $k_V=-20.75$.

To fit the B-V color index, use the same extinction law as:

$$(B-V)^A=(B-V)+(k_B-k_V) X$$

where we can find:
- $(B-V)^A=m^A_B-m^A_V$
- B-V from the SIMBAD database [5]
- the airmass $X$ from previous calculations
- and the difference between the extinction in the B and V filters $(k_B-k_V)$ by fitting.

```python
df['B-V_instr']=df['m_instr_B']-df['m_instr_V']
df['B-V_diff']=df['B-V_instr']-df['B-V']
df
```

```python
#performing fitting using the same function y (linear fit)

popt2, pcov2 = curve_fit(y, df['X'], df['B-V_diff'])
```

```python
k_B_V = popt2[0]
print('Slope: k=', k_B_V)
```

```python
#plot to illustrate
x_fit = np.linspace(min(df['X']), max(df['X']), 100)  # 100 points from min(x) to max(x) for a smooth line
y_fit = y(x_fit, k_B_V)

plt.scatter(df['X'], df['B-V_diff'], label = 'Data', color='black', marker='+')
plt.plot(x_fit, y_fit, label = 'Fitted line', color='red', ls='--')
#plt.xlim(0.999, 1.001)
plt.title(r'Fitting to find $(k_B-k_V)$')
plt.xlabel(r'Airmass $X$')
plt.ylabel(r'$(B-V)^A-(B-V)$')
plt.legend()
```

**Result 2:** we have found the difference between the extinction in the B and V filters to be $(k_B-k_V)=-1.24$.

Now, we transition to the Johnson photometric system to obtain the required coefficients:

$$(m^A_V-m_V)=\beta+\gamma(B-V)$$
$$(B-V)^A=\eta+\epsilon(B-V)$$

where:
- $(B-V)$ is the colour index outside the atmosphere, obtained from SIMBAD
- $(B-V)^A$ is the colour index at the instrument, after atmospheric effects
- $m^A_V$ is the magnitude in the V band at the instrument, after atmospheric effects
- $m_V$ is the magnitude in the V band without any atmospheric effects, obtained from SIMBAD
- and $\beta, \gamma, \eta, \epsilon$ are constants which will be obtained from fitting.

```python
#df.head()
```

```python
#define a new relation y=mx+c with 2 parameters to fit
def y2(x, m, c):
    return m*x+c
```

```python
popt3, pcov3 = curve_fit(y2, df['B-V'], df['m_instr_V-m_V'])
```

```python
gamma, beta = popt3
print('gamma (gradient) = ', gamma, '\nbeta (intercept) = ', beta)
```

```python
#plot to illustrate
x_fit = np.linspace(min(df['B-V']), max(df['B-V']), 100)  # 100 points from min(x) to max(x) for a smooth line
y_fit = y2(x_fit, gamma, beta)

plt.scatter(df['B-V'], df['m_instr_V-m_V'], label = 'Data', color='black', marker='+')
plt.plot(x_fit, y_fit, label = 'Fitted line', color='red', ls='--')
#plt.xlim(0.999, 1.001)
plt.title(r'Fitting to find $\gamma$ and $\beta$')
plt.xlabel(r'$B-V$')
plt.ylabel(r'$m^A_V-m_V$')
plt.legend()
```

```python
#df.head()
```

```python
popt4, pcov4 = curve_fit(y2, df['B-V'], df['B-V_instr'])
```

```python
epsilon, eta = popt4
print('epsilon (gradient) = ', epsilon, '\neta (intercept) = ', eta)
```

```python
#plot to illustrate
x_fit = np.linspace(min(df['B-V']), max(df['B-V']), 100)  # 100 points from min(x) to max(x) for a smooth line
y_fit = y2(x_fit, epsilon, eta)

plt.scatter(df['B-V'], df['B-V_instr'], label = 'Data', color='black', marker='+')
plt.plot(x_fit, y_fit, label = 'Fitted line', color='red', ls='--')
#plt.xlim(0.999, 1.001)
plt.title(r'Fitting to find $\epsilon$ and $\eta$')
plt.xlabel(r'$B-V$')
plt.ylabel(r'$(B-V)^A$')
plt.legend()
```

**Result 3:** our final results for the constants are: $\gamma=0.039$ and $\beta=-20.786$ from the first equation, and $\epsilon=0.62$ and $\eta=-1.03$ from the second equation.

Next, we correct the observed magnitude of the target star using the above derived relations.

```python
#data for target star which needs correction
target_ra = (19*60)+34.1 #alpha = right ascencion (in minutes)
target_dec = (31*60)+18 #delta = declination (in arcmin)

target_N_V = 3.56e5 #no. of photons in V-band
target_N_B = 7.7e5 #no. of photons in B-band
target_LST = (19*60)+52 #local sidereal time in minutes
```

Calculating the instrumental (observed) magnitudes $m^A_B$ and $m^A_V$ from the received photon counts using the same relation as before:

$$m^A_\lambda = -2.5 \log(N_\lambda)$$

```python
target_m_instr_V=-2.5*np.log10(target_N_V)
target_m_instr_B=-2.5*np.log10(target_N_B)
```

To find $Z$, we need to find the coordinates of the star by computing the Hour Angle:

$$H = \mathrm{LST} - \alpha$$

Note that currently, both the LST and $\alpha$ are in minutes.

```python
target_HA = target_LST - target_ra #find hour angle in minutes
#check result
print('Hour angle [mins] = ', target_HA)
```

The result lies within [0,24), so we do not need to account for negative values.

Now, we need to convert the hour angle into degrees. Following the same procedure as before:

1 Hour Angle [hours] = 60 Hour Angle [minutes] = $15\degree$

such that 1 Hour Angle [minutes] = $15/60\degree$.

```python
target_HA_deg = target_HA*15/60
```

Next, the zenith distance $Z$ can be found from: $$Z=\arccos(\sin\phi\sin\delta+\cos\phi\cos\delta\cos H)$$

```python
target_Z=find_Z(obs_phi, target_dec, target_HA_deg)
```

Now we can find the air mass X as before from:
$$X=\dfrac{1}{\cos(Z)}\left(1-0.0012\left(\dfrac{1}{\cos^2(Z)}-1\right)\right)$$

```python
target_X = find_X(target_Z)
print(target_X)
```

Using the above derived information, we now find the magnitude before extinction in the V band using:
$$m^A_\lambda=m_\lambda+k_\lambda X$$

```python
target_m_V = target_m_instr_V-k_V*target_X
print('target star m_V = ', target_m_V)
```

To find the target star's B-V colour index:

$$(B-V)^A=(B-V)+(k_B-k_V) X$$

```python
target_m_instr_B_V = target_m_instr_B - target_m_instr_V
target_m_B_V = target_m_instr_B_V - k_B_V * target_X
print('target star B-V = ', target_m_B_V)
```

Transitioning to the Johnson photometric system to find the magnitudes using an alternative method:

$$(m^A_V-m_V)=\beta+\gamma(B-V)$$
$$(B-V)^A=\eta+\epsilon(B-V)$$

$$\implies(B-V)=\dfrac{(B-V)^A-\eta}{\epsilon}~\mathrm{and}~m_V=m^A_V-\beta-\gamma(B-V)$$

$$\mathrm{combining}~\implies m_V=m^A_V-\beta-\gamma\left(\dfrac{(B-V)^A-\eta}{\epsilon}\right)$$

```python
star_B_V = (target_m_instr_B_V-eta)/epsilon
print(r'B-V of target star using 2nd method = ', star_B_V)
```

```python
star_m_V = target_m_instr_V-beta-gamma*star_B_V
print(r'm_V of target star using 2nd method = ', star_m_V)
```

**Result 4:** Both methods give very close estimates. The corrected magnitude of the target star in the V band is around $m_V=6.90$, and the corrected colour index of the target star is around $(B-V)=0.32$.
